```python
import pandas as pd
import matplotlib.pyplot as plt
import shap
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

pd.set_option('display.max_rows',150)
pd.set_option('display.max_columns',100) 
```

```python
df=pd.read_csv("R3_DATA.csv")


prod_map={'Product 1': 1, 'Product 2': 2, 'Product 3': 3, 'Product 4': 4, 'Product 5': 5, 'Product 6': 6, 'Product 7': 7, 'Product 8': 8, 'Product 9': 9, 'Product 10': 10, 'Product 11': 11, 'Product 12': 12}
channel_map={'Channel 1': 1, 'Channel 2': 2, 'Channel 3': 3, 'Channel 4': 4}
bureau_src={'No Bureau Hit':0, 
            'Bureau 1':1,
            'Bureau 2':2}
yes_no_trust = {np.nan:1,
          "N": 0,
          "Y": 1}
yes_no_addrs = {np.nan:0,
          "N": 0,
          "Y": 1}

from sklearn.preprocessing import LabelEncoder

months_array = np.array(['Feb', 'Jun', 'May', 'Mar', 'Apr', 'Jan'])
label_encoder = LabelEncoder()
label_encoder.fit(months_array)
df["appl_month"] = label_encoder.transform(df["appl_month"])

df["prod_name"] = df["prod_name"].map(prod_map)
df["acq_channel"] = df["acq_channel"].map(channel_map)

label_encoder = LabelEncoder()
df["state_code"] = label_encoder.fit_transform(df["state_code"])

df['bureau_src'] = df['bureau_src'].map(bureau_src)
df["addr_mismatch"] =df["addr_mismatch"].map(yes_no_addrs)
df["trust_identity"] = df["trust_identity"].map(yes_no_trust)

weights=[0.0, 0.0010573924150949557, 0.18989355413166095, 0.1775521434473567, 0.08908306519691163, 0.11896316724878306, 0.0774301589391904, 0.15063919774330004, 0.0011355634978398293, 0.027198420032897564, 0.0010573924150949557, 0.04963654700162538]
df["Risk_Total"]=0
for j in range (1,12):
    scaler = StandardScaler()
    df[f"risk_score_{j}"] = scaler.fit_transform(df[[f"risk_score_{j}"]])
    df["Risk_Total"]+=weights[j]*df[f"risk_score_{j}"]
    if j in [10,6,5]:
        df["Risk_Total"]-=2*weights[j]*df[f"risk_score_{j}"]
    df=df.drop([f"risk_score_{j}"],axis=1)

df['merchant_list'] = df.apply(lambda row: [row['merchant1'], row['merchant2'], row['merchant3']], axis=1)

df["percentage_decline"] = np.where(df["attempt_txn"] == 0, -1, df["decline_txn"] * 100 / df["attempt_txn"])
df["percentage_outside"] = np.where(df["no_trd"] == 0, -1, df["no_open_trd"] * 100 / df["no_trd"])


# df=df.drop(["merchant1","merchant1_ten","merchant1_cat","merchant1_amt",
#             "merchant2","merchant2_ten","merchant2_cat","merchant2_amt",
#             "merchant3","merchant3_ten","merchant3_cat","merchant3_amt",
#             "state_code","bureau_src","spend","payments",
#            "decline_txn","attempt_txn","uid_addr_matches","uid_nid_matches","trust_identity"],axis=1)

fraud=df[df["default_ind"]==1]
real=df[df["default_ind"]==0]
```

```python
def get_month_fraud_merchant(df1,month):
    df=df1[df1["appl_month"]==month]
    all_merchant_ids = [merchant_id for sublist in df['merchant_list'] for merchant_id in sublist]
    id_counts = pd.Series(all_merchant_ids).value_counts()
    # Plotting
    plt.figure(figsize=(128, 6))
    sns.barplot(x=id_counts.index, y=id_counts.values, palette='viridis')
    plt.xlabel('Merchant ID')
    plt.ylabel('Count')
    plt.title('Count of Each Merchant ID')
    plt.show()
for j in range (0,6):
    get_month_fraud_merchant(fraud,j)
```

```python
month_fraud= fraud.sort_values(by="appl_month")
```

```python
def get_fraud_using_merchant(df):
    fraud_merchant = list(set([100002,124596,156589,163070,164843,115459,110104,164428,164845,167574]))
    df_filtered = df[df['merchant_list'].apply(lambda merchants: sum(merchant in fraud_merchant for merchant in merchants) >= 1)]
    return df_filtered
```

```python
real_fraud=get_fraud_using_merchant(month_fraud)
```

```python
real_fraud = real_fraud[(real_fraud["prod_name"] == 1) & 
                        (real_fraud["acq_channel"] == 3) & 
                        ((real_fraud["bureau_mismatch"] >= 1) | 
                         (real_fraud["addr_mismatch"] >= 1) | 
                         (real_fraud["no_open_trd"] != 0)) & 
                        (real_fraud["income"] < 100000)]
real_fraud = real_fraud[real_fraud["appl_month"].isin([0,4,5])]
```

```python
def plot_count(col):
    target_sum = df.groupby(col)['default_ind'].sum().reset_index()
    average_sum = target_sum['default_ind'].mean()
    plt.figure(figsize=(8, 5))
    sns.barplot(x=col, y='default_ind', data=target_sum, color='blue')
    plt.axhline(average_sum, color='red', linestyle='--', label=f'Average: {average_sum:.2f}')
    
    plt.xlabel('Column of Interest')
    plt.ylabel('Sum of Target (1)')
    plt.title('Sum of Target per Column Value')
    plt.show()
```

```python
def make_plot_continues(col):
    bin_labels = pd.cut(df[col], bins=40)
    grouped = df.groupby(bin_labels)["default_ind"].sum().reset_index()
    plt.figure(figsize=(10, 6))
    plt.bar(grouped[col].astype(str), grouped["default_ind"], color='skyblue', edgecolor='black')
    plt.xlabel('Ranges of Continuous Column')
    plt.ylabel('Sum of Target Values')
    plt.title('Sum of Target Values for Each Range of Continuous Column')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
